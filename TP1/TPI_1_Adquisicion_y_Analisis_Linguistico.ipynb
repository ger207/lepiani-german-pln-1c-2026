```python
# 1.1 Scraping en vivo
urls = [
    "https://www.infobae.com/deportes/2026/04/14/barcelona-atletico-de-madrid-y-liverpool-psg-definiran-los-primeros-dos-semifinalistas-de-la-champions-league-todo-lo-que-hay-que-saber/"
]
def extraer_noticias_web(urls):
    """Extrae el texto de una lista de URLs usando Trafilatura"""
    noticias = []
    for url in urls:
        try:
            downloaded = trafilatura.fetch_url(url)
            text = trafilatura.extract(downloaded)
            if text:
                noticias.append({'titulo_o_fuente': url, 'texto': text, 'origen': 'web'})
        except Exception as e:
            print(f"Error al procesar {url}: {e}")
    return noticias
```

```python
import yt_dlp

url_video = "https://www.youtube.com/watch?v=74kKGcWc2H8"  # Ejemplo de video

def transcribir_audio_youtube(url_video):
    """Descarga audio de YouTube y transcribe con Whisper."""
    if isinstance(url_video, str):
        url_video = [url_video]

    model = whisper.load_model("small", device="cpu")
    resultados = []

    ydl_opts = {
        "format": "bestaudio/best",
        "outtmpl": "temp_audio.%(ext)s",
        "noplaylist": True,
        "quiet": True,
        "no_warnings": True,
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        for url in url_video:
            try:
                info = ydl.extract_info(url, download=True)
                audio_path = ydl.prepare_filename(info)

                result = model.transcribe(audio_path, language="es")
                resultados.append({
                    "titulo_o_fuente": url,
                    "texto": result["text"],
                    "origen": "audio"
                })
            except Exception as e:
                print(f"Error transcribiendo {url}: {e}")
            finally:
                if os.path.exists(audio_path):
                    os.remove(audio_path)

    return resultados
```

```python
import yt_dlp
from datetime import datetime

url_video = "https://www.youtube.com/watch?v=74kKGcWc2H8"

_MODELOS_WHISPER = {}

def transcribir_audio_youtube(url_video):
    
    try:
        carpeta_audios = "audios"
        os.makedirs(carpeta_audios, exist_ok=True)

        opciones = {
            "format": "bestaudio/best",
            "noplaylist": True,
            "outtmpl": os.path.join(carpeta_audios, "audio.%(ext)s"),
            "quiet": True,
            "no_warnings": True
        }

        with yt_dlp.YoutubeDL(opciones) as ydl:
            info = ydl.extract_info(url_video, download=True)

        titulo = info.get("title", url_video)
        fecha_publicacion = info.get("upload_date")
        if fecha_publicacion:
            fecha_publicacion = f"{fecha_publicacion[:4]}-{fecha_publicacion[4:6]}-{fecha_publicacion[6:8]}"

        archivo_audio = None
        for archivo in os.listdir(carpeta_audios):
            if archivo.startswith("audio"):
                archivo_audio = os.path.join(carpeta_audios, archivo)
                break

        if archivo_audio is None:
            print("No se encontró el audio descargado")
            return []

        if "base" not in _MODELOS_WHISPER:
            _MODELOS_WHISPER["base"] = whisper.load_model("base", device="cpu")

        modelo = _MODELOS_WHISPER["base"]
        resultado = modelo.transcribe(archivo_audio, language="es", fp16=False)

        registro_audio = {
            "titulo_o_fuente": titulo,
            "texto": resultado["text"],
            "origen": "audio",
            "tipo": "transcripcion",
            "fecha_extraccion": datetime.now().strftime("%Y-%m-%d"),
            "fecha_publicacion": fecha_publicacion,
            "url": url_video
        }

        return [registro_audio]

    except Exception as e:
        print(f"Error transcribiendo audio: {e}")
        return []
```

```python
datos_audio = transcribir_audio_youtube(url_video)
df_audio = pd.DataFrame(datos_audio)
print(df_audio.head())
```

```python
# 1.3 Carga de JSON local
def cargar_json_previo(ruta_json):
    """Carga un corpus pre-extraído en formato JSON"""
    if ruta_json and os.path.exists(ruta_json):
        with open(ruta_json, 'r', encoding='utf-8') as f:
            data = json.load(f)
        return data
    return []
```

```python
# 1.4 Consolidación
def unificar_corpus(datos_web, datos_audio, datos_json):
    """Unifica las tres fuentes en un DataFrame con columnas estándar"""
    all_data = datos_web + datos_audio + datos_json
    df_unificado = pd.DataFrame(all_data)
    return df_unificado

# ---- Espacio para pruebas ----
# Probá tus funciones acá con al menos 1 url web y 1 video corto.
urls_prueba = ["https://www.infobae.com/deportes/2026/04/14/barcelona-atletico-de-madrid-y-liverpool-psg-definiran-los-primeros-dos-semifinalistas-de-la-champions-league-todo-lo-que-hay-que-saber/"]  # URL de ejemplo (elige una real y corta)
audio_prueba = ["https://www.youtube.com/watch?v=74kKGcWc2H8"]
json_prueba = []  # O una ruta si tienes un JSON

datos_web = extraer_noticias_web(urls_prueba)
datos_audio = transcribir_audio_youtube(audio_prueba) if audio_prueba else []
datos_json = cargar_json_previo(json_prueba) if json_prueba else []

df_corpus = unificar_corpus(datos_web, datos_audio, datos_json)
print("DataFrame unificado:")
print(df_corpus.head())  # Para ver el DataFrame resultante
```

## Parte 2: Análisis Lingüístico con spaCy

**Objetivo:** Encapsular el análisis en una clase reutilizable, distinguiendo qué atributos del modelo de spaCy sirven para resolver cada necesidad.

> [!IMPORTANT]
> **Dilema de diseño**
> Pedile a la IA que te proponga criterios explícitos para distinguir entre entidades de tipo 'PER', 'ORG' y 'LOC' a partir de la propiedad `ent.label_` de spaCy. Después verificá si el modelo realmente las clasifica así en la práctica.
> Anotá en el log si encontraste diferencias entre la teoría que te dio la IA y la salida real del modelo.

```python
class AnalizadorCorpus:
    def __init__(self, df, modelo_spacy="es_core_news_lg"):
        self.df = df
        print("Cargando modelo de lenguaje...")
        self.nlp = spacy.load(modelo_spacy)
        
        # Procesamos la columna 'texto' al instanciar la clase
        print("Procesando los textos con spaCy...")
        self.df['doc'] = self.df['texto'].apply(self.nlp)
        
    def extraer_entidades(self):
        """Devuelve las entidades agrupadas por tipo, contabilizando total de apariciones"""
        entidades = {}
        for doc in self.df['doc']:
            for ent in doc.ents:
                tipo = ent.label_
                texto = ent.text
                if tipo not in entidades:
                    entidades[tipo] = {}
                if texto not in entidades[tipo]:
                    entidades[tipo][texto] = 0
                entidades[tipo][texto] += 1
        return entidades

    def extraer_verbos_principales(self, n=15):
        """Devuelve los 'n' verbos lematizados más frecuentes de todo el corpus"""
        verbos = Counter()
        for doc in self.df['doc']:
            for token in doc:
                if token.pos_ == 'VERB' and not token.is_stop:
                    verbos[token.lemma_] += 1
        return verbos.most_common(n)

    def extraer_palabras_clave(self, n=20):
        """Devuelve sustantivos y nombres propios lematizados y filtrados"""
        palabras = Counter()
        for doc in self.df['doc']:
            for token in doc:
                if token.pos_ in ['NOUN', 'PROPN', 'ADJ'] and not token.is_stop and not token.is_punct:
                    palabras[token.lemma_] += 1
        return palabras.most_common(n)
        
    def estadisticas_corpus(self):
        """Genera un diccionario con métricas generales del corpus"""
        total_tokens = sum(len(doc) for doc in self.df['doc'])
        vocab = set()
        for doc in self.df['doc']:
            for token in doc:
                vocab.add(token.lemma_)
        vocab_size = len(vocab)
        total_sentences = sum(len(list(doc.sents)) for doc in self.df['doc'])
        return {'total_tokens': total_tokens, 'vocab_size': vocab_size, 'total_sentences': total_sentences}
```

## Parte 4: Pipeline Integrado (Orquestación)

**Objetivo:** Orquestar los componentes desarrollados en un flujo lógico unificado y persistir los hallazgos en formato estructurado. Todo sistema analítico debe poder guardar su estado final de forma estática.

```python
class PipelineMediatico:
    def __init__(self, urls_web=None, url_audio=None, ruta_json=None):
        self.urls_web = urls_web or []
        self.url_audio = url_audio
        self.ruta_json = ruta_json
        self.df = None
        self.analizador = None
        
    def ejecutar_pipeline(self):
        """Orquesta la adquisición, unificación y análisis"""
        datos_web = extraer_noticias_web(self.urls_web)
        datos_audio = transcribir_audio_youtube(self.url_audio) if self.url_audio else []
        datos_json = cargar_json_previo(self.ruta_json) if self.ruta_json else []
        self.df = unificar_corpus(datos_web, datos_audio, datos_json)
        if not self.df.empty:
            self.analizador = AnalizadorCorpus(self.df)
        print("Pipeline ejecutado exitosamente.")
        
    def generar_reporte_y_exportar(self, ruta_csv="corpus_resultante.csv", ruta_json="estadisticas.json"):
        """Exporta el dataframe y un JSON analítico"""
        if self.df is not None:
            df_export = self.df.drop(columns=['doc'], errors='ignore')
            df_export.to_csv(ruta_csv, index=False)
        if self.analizador is not None:
            stats = self.analizador.estadisticas_corpus()
            entidades = self.analizador.extraer_entidades()
            with open(ruta_json, 'w', encoding='utf-8') as f:
                json.dump({'estadisticas': stats, 'entidades': entidades}, f, ensure_ascii=False, indent=4)

# ---- Espacio para pruebas ----
# pipeline = PipelineMediatico(urls_web=["..."], url_audio="...")
# pipeline.ejecutar_pipeline()
# pipeline.generar_reporte_y_exportar()
# Para probar con datos de ejemplo
pipeline = PipelineMediatico(urls_web=["https://www.infobae.com/deportes/2026/04/14/barcelona-atletico-de-madrid-y-liverpool-psg-definiran-los-primeros-dos-semifinalistas-de-la-champions-league-todo-lo-que-hay-que-saber/"], url_audio=None)  # Deshabilitado para evitar descarga
pipeline.ejecutar_pipeline()
pipeline.generar_reporte_y_exportar()
print("Pipeline completado. Archivos exportados si el pipeline funcionó.")
```
